```python
# Importing file.
import pandas as pd

# File to load.
vine_table_to_load = "vine_table.csv"

# Read in the file.
vine_table_df = pd.read_csv(vine_table_to_load)
vine_table_df.head()
```

```python
# Filter data for votes.
vine_votes = vine_table_df.loc[(vine_table_df["total_votes"] >= 20)]
vine_votes
```

```python
vine_prop = vine_votes.loc[((vine_votes["helpful_votes"] / vine_votes["total_votes"]) >= .5)]
vine_prop
```

```python
# filterting for paid reviews.
vine_paid = vine_prop.loc[(vine_prop["vine"] == "Y")]
vine_paid
```

```python
# filtering for unpaid reviews
vine_unpaid = vine_prop.loc[(vine_prop['vine'] == "N")]
vine_unpaid
```

```python
# Total Reviews paid.
vine_paid.count()
```

```python
# Total Reviews unpaid.
vine_unpaid_count = vine_unpaid["review_id"].count()
vine_unpaid_count
```

```python
# Total 5 Star reviews.
vine_great_reviews = vine_unpaid.loc[vine_unpaid["star_rating"] == 5]["star_rating"].count()
vine_great_reviews
```

```python
# Percentage of 5 star reviews.
vine_great_prop = vine_great_reviews / float(vine_unpaid_count) * 100
vine_great_prop
```
